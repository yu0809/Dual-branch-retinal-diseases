```python
import pandas as pd  
  
def is_consecutive(lst):  
    """检查列表中的数字是否连续"""  
    return lst == list(range(min(lst), max(lst) + 1))  
  
def extract_numbers_from_filenames(filenames):  
    """从文件名中提取数字部分"""  
    return [int(filename.split('.')[0]) for filename in filenames]  
  
def check_consecutive_filenames(csv_file):  
    # 读取CSV文件  
    df = pd.read_csv(csv_file)  
      
    # 提取filename列  
    filenames = df['filename'].tolist()  
      
    # 从文件名中提取数字  
    numbers = extract_numbers_from_filenames(filenames)  
      
    # 检查数字是否连续  
    consecutive = is_consecutive(numbers)  
      
    return consecutive  
  
# 使用示例  
csv_file = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\最终数据处理.csv'  # 请替换为你的CSV文件路径  
consecutive = check_consecutive_filenames(csv_file)  
print(f"The filenames are {'consecutive' if consecutive else 'not consecutive'}.")
```

```python
import os

def extract_jpg_filenames(folder_path):
    # 创建一个列表来存储找到的文件名
    filenames = []
    
    # 遍历文件夹中的所有文件
    for filename in os.listdir(folder_path):
        # 检查文件名是否以.jpg结尾
        if filename.endswith(".jpg"):
            # 尝试将文件名（不包括扩展名）转换为整数
            try:
                file_number = int(filename.split('.')[0])
                # 如果文件名在指定的范围内，则添加到列表中
                if 1 <= file_number <= 6392:
                    filenames.append(filename)
            except ValueError:
                # 如果转换失败，说明文件名不是纯数字，忽略该文件
                continue
                
    return filenames

# 使用示例
folder_path = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\椭圆的'  # 将'your/folder/path'替换为你的文件夹路径
filenames = extract_jpg_filenames(folder_path)
print(filenames)
```

```python
import pandas as pd  
  
# 读取CSV文件  
file_path = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\four_labels_data.csv'  # 替换为你的CSV文件路径  
df = pd.read_csv(file_path)  
  
# 如果labels列的值是列表的字符串表示（例如："['C']"），则需要先将它们转换为单个标签字符串  
# 假设labels列的值是类似"['C']"这样的格式，我们需要去掉方括号和引号  
import ast  
  
def clean_label(label):  
    # 使用ast.literal_eval将字符串表示的列表转换回列表  
    label_list = ast.literal_eval(label)  
    # 假设列表只有一个元素，我们取第一个元素作为标签  
    return label_list[0] if label_list else None  
  
# 应用clean_label函数到labels列  
df['labels'] = df['labels'].apply(clean_label)  
  
# 现在统计labels列中各个值的出现次数  
label_counts = df['labels'].value_counts()  
  
# 打印结果  
print(label_counts)
```

```python
import pandas as pd  
import os  
import shutil  
  
# 读取CSV文件  
csv_file_path = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\modified_four_labels_data.csv'  # CSV文件路径  
df = pd.read_csv(csv_file_path)  
  
# 图片所在文件夹路径和新文件夹路径  
source_folder = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\preprocessed_images'  # 原始图片文件夹路径  
destination_folder = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\unturned_images'  # 目标文件夹路径  
  
# 确保目标文件夹存在  
if not os.path.exists(destination_folder):  
    os.makedirs(destination_folder)  
  
# 遍历DataFrame，根据labels列的值复制对应的图片  
for index, row in df.iterrows():  
    label = row['filename']  # 获取labels列的值  
    filename = f"{label}.jpg"  # 构造对应的图片文件名  
    source_path = os.path.join(source_folder, filename)  # 原始图片完整路径  
    destination_path = os.path.join(destination_folder, filename)  # 目标图片完整路径  
      
    # 检查源文件是否存在，如果存在则复制到目标文件夹  
    if os.path.exists(source_path):  
        shutil.copy2(source_path, destination_path)  # 复制文件（包括元数据）  
        print(f"Copied {filename} to {destination_folder}")  
    else:  
        print(f"Warning: {filename} not found in {source_folder}")  
  
print("All matching images have been copied to the destination folder.")
```

```python
import os  
  
# 设置文件夹路径  
source_folder = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\unturned_images'  # 当前图片所在的文件夹路径  
  
# 获取文件夹中所有文件的名字和序号  
files = os.listdir(source_folder)  
file_with_numbers = [(file, int(file.split('.')[0])) for file in files if file.endswith('.jpg')]  
  
# 根据序号排序  
sorted_files = sorted(file_with_numbers, key=lambda x: x[1])  
  
# 重新命名文件，赋予新的连续序号  
for i, (old_name, old_number) in enumerate(sorted_files, start=1):  
    # 构造新的文件名  
    new_name = f"{i}.jpg"  
    # 完整的旧文件路径和新文件路径  
    old_path = os.path.join(source_folder, old_name)  
    new_path = os.path.join(source_folder, new_name)  
    # 重命名文件  
    os.rename(old_path, new_path)  
  
print("All files have been renamed with consecutive numbers.")
```

```python
import os  
from PIL import Image  
  
# 设置原文件夹路径和新文件夹路径  
source_folder = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\unturned_images'  # 原始图片所在的文件夹路径  
destination_folder = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\turned_images'  # 镜像翻转后图片保存的文件夹路径  
  
# 确保新文件夹存在  
if not os.path.exists(destination_folder):  
    os.makedirs(destination_folder)  
  
# 遍历原文件夹中的所有.jpg文件  
for filename in os.listdir(source_folder):  
    if filename.endswith('.jpg'):  
        # 构造完整的文件路径  
        source_path = os.path.join(source_folder, filename)  
        # 提取文件序号  
        file_number = int(filename.split('.')[0])  
        # 计算新的文件名序号  
        new_file_number = file_number + 4899  
        # 构造新的文件名和路径  
        new_filename = f"{new_file_number}.jpg"  
        destination_path = os.path.join(destination_folder, new_filename)  
          
        # 打开图片并进行镜像翻转  
        with Image.open(source_path) as img:  
            mirrored_img = img.transpose(Image.FLIP_LEFT_RIGHT)  
            # 保存镜像翻转后的图片到新文件夹  
            mirrored_img.save(destination_path)  
  
print("All images have been mirrored and saved to the new folder.")
```

```python
import os  
import pandas as pd  
import shutil  
  
# CSV文件路径  
csv_file_path = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\modified_four_labels_data.csv'  
# 图片所在文件夹路径  
images_folder_path = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\unturned_images'  
  
# 读取CSV文件  
df = pd.read_csv(csv_file_path)  
  
# 为每个标签创建一个文件夹  
label_folders = {'C': 'folder_C', 'D': 'folder_D', 'M': 'folder_M', 'N': 'folder_N'}  
for label in label_folders.values():  
    os.makedirs(os.path.join(images_folder_path, label), exist_ok=True)  
  
# 遍历DataFrame，根据labels列的值移动对应的图片到相应文件夹  
for index, row in df.iterrows():  
    filename = f"{row['filename']}.jpg"  # 假设所有图片都是.jpg格式  
    # 去掉标签字符串中的方括号  
    label_str = row['labels'].strip('[]')  
    # 如果标签字符串中包含单引号或双引号，也去掉它们  
    label = label_str.strip("'\"")  
      
    # 构造源文件和目标文件的完整路径  
    source_path = os.path.join(images_folder_path, filename)  
    destination_folder = os.path.join(images_folder_path, label_folders[label])  
    destination_path = os.path.join(destination_folder, filename)  
      
    # 检查源文件是否存在，如果存在则移动到目标文件夹  
    if os.path.exists(source_path):  
        shutil.move(source_path, destination_path)  
        print(f"Moved {filename} to {destination_folder}")  
    else:  
        print(f"Warning: {filename} not found in {images_folder_path}")  
  
print("All images have been moved to their respective folders based on labels.")
```
