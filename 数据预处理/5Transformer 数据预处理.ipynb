```python
import numpy as np
import os
from skimage.transform import resize
from tqdm import tqdm

# 定义一个函数来调整图像尺寸并保存
def resize_and_save_npy(input_dir, output_dir, new_size=(224, 224)):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for filename in tqdm(os.listdir(input_dir)):
        if filename.endswith('.npy'):
            file_path = os.path.join(input_dir, filename)
            image = np.load(file_path)

            # 确保图像是float类型，用于resize函数
            image_resized = resize(image, new_size, anti_aliasing=True, preserve_range=True)
            
            # 直接保存调整尺寸后的float32数据
            output_file_path = os.path.join(output_dir, filename)
            np.save(output_file_path, image_resized)

# 输入和输出目录
input_dir = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\segmentation\normalized_predicted_train'
output_dir = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\segmentation\transformer_predicted_train'

# 调整图像尺寸并保存
resize_and_save_npy(input_dir, output_dir)
```

```python
import numpy as np
import os

def normalize_npy_files(input_dir, output_dir):
    """
    将指定目录下的所有.npy文件归一化到[0, 1]范围内，并保存到输出目录。
    
    :param input_dir: 原始.npy文件的目录
    :param output_dir: 归一化后的.npy文件的保存目录
    """
    # 检查输出目录是否存在，如果不存在则创建
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 遍历输入目录中的所有文件
    for filename in os.listdir(input_dir):
        if filename.endswith('.npy'):
            file_path = os.path.join(input_dir, filename)
            # 加载.npy文件
            image = np.load(file_path)

            # 计算当前图像的最小值和最大值
            image_min = image.min()
            image_max = image.max()

            # 归一化到[0, 1]
            image_normalized = (image - image_min) / (image_max - image_min)

            # 构造输出文件的完整路径
            output_file_path = os.path.join(output_dir, filename)

            # 保存归一化后的.npy文件
            np.save(output_file_path, image_normalized)
            print(f'Processed and saved: {output_file_path}')

# 输入和输出目录
input_dir = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\segmentation\transformer_predicted_validation'  # 请替换为您的原始.npy文件目录
output_dir = r'C:\Users\HP\Desktop\指南者\项目\眼病预测\segmentation\transformer_predicted_validation'  # 请替换为您希望保存归一化文件的目录

# 调用函数处理所有.npy文件
normalize_npy_files(input_dir, output_dir)
```
